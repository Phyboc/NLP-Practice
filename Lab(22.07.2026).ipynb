2.Write a python code with NLTK to print the tokens and stop words.

```python
import nltk
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
def get_tokens():
    text = """I am Sivasubramani K J, a Bachelor of Technology student in Computer Science and Engineering at 
    Amrita Vishwa Vidyapeetham, Coimbatore. My academic interests include computer science fundamentals, 
    software development, artificial intelligence, machine learning, embedded systems, and data-driven 
    technologies. I have gained experience in programming languages such as Java, Python, C, and C++, along 
    with exposure to various technical domains through coursework and academic projects. I continuously strive 
    to strengthen my knowledge and understanding of emerging technologies in computer science."""
    lowers = text.lower()
    no_punctuation = lowers.translate(str.maketrans("", "", string.punctuation))
    tokens = word_tokenize(no_punctuation)
    return tokens

def get_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    stopWordslst = [word for word in tokens if word.lower() in stop_words]
    return stopWordslst

tokens = get_tokens()
stopWords = get_stopwords(tokens)
count = Counter(tokens)
print("Most common words: ")
print(count.most_common(10))

print("Stop Words: ")
print(stopWords)

print("All of the words and count")
print(tokens)
```

3. Write a set of sentences about the food you like. Conclude what type of foodie you are.
- "I like Aloo Parotta and Bun Parotta along with veg kuruma. I also like to eat sweet food items. Sometimes, I do like packed food items too. Ultimately, I am an eclectic, comfort-driven foodie who loves exploring both bold local spices and hearty, globally-inspired classics."


Implement using NLTK to print the phases in NLP.

```python
import nltk
from nltk import sent_tokenize, word_tokenize, pos_tag
nltk.download('averaged_perceptron_tagger_eng', quiet=True)


text = """I like Aloo Parotta and Bun Parotta along with veg kuruma. I also like to eat sweet food items.
 Sometimes, I do like packed food items too. Ultimately, I am an eclectic, 
 comfort-driven foodie who loves exploring both bold local spices and hearty, globally-inspired classics. """
sentences = sent_tokenize(text)
print("--- PHASE 1: SENTENCE TOKENIZATION ---")
for i, sent in enumerate(sentences):
    print(f"Sentence {i + 1}: {sent}\n")

words = word_tokenize(text)
print("--- PHASE 2: WORD TOKENIZATION ---")
print(words, "\n")

pos_tags = nltk.pos_tag(words)
print("--- PHASE 3: PART-OF-SPEECH (POS) TAGGING ---")
print(pos_tags)
```
